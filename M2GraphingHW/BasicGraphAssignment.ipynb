Graph 1: Opioids in VA

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
df_opiva = pd.read_csv('OpidsVA.csv')
```

```python
fig, ax = plt.subplots()
ax.scatter(df_opiva['MedianHouseholdIncome'], df_opiva['FPOO-Rate'], alpha=0.5, c = 'b')
fig.suptitle('Opioid Overdoses vs. Median Household Income')
ax.xaxis.set_label_text('Median Household Income')
ax.yaxis.set_label_text('Opioid Overdoses per 100K Residents')
#ax.set_xlim(0,100000)  # Set min and max for x axis
#ax.set_ylim(0,50)  # Set min and max for y axis 
fig.set_size_inches(7,5)
plt.show()

#I chose a scatter plot because scatter plots are best for comparing two variables.
```

Yes, opioid overdoses tend to be associated with less affluent areas.

```python
fig, ax = plt.subplots()
ax.scatter(df_opiva['FPOO-Rate'], df_opiva['FFHO-Rate'], alpha=0.5, c = 'b')
fig.suptitle('Opioid Overdoses vs. Heroin Overdoses')
ax.xaxis.set_label_text('Opioid Overdoses per 100K Residents')
ax.yaxis.set_label_text('Heroin Overdoses per 100K Residents')
#ax.set_xlim(0,100000)  # Set min and max for x axis
#ax.set_ylim(0,50)  # Set min and max for y axis 
fig.set_size_inches(7,5)
plt.show()

#I chose a scatter plot because scatter plots are best for comparing two variables.
```

There does not appear to be a strong correlation between opioid overdoes and heroin overdoses, the data shows people likely experience one or the other. 

Graph 2: U.S. Presidents by Home State

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
df_pres = pd.read_csv('presidents.csv')
```

```python
df_pres['State'].value_counts().plot(kind='bar') 
plt.ylabel("No. of Presidents")
plt.title("List of U.S. Presidents by Birth State")
plt.xlabel("Birth State")

# I chose a bar chart because it easiest for graphing the count objects instead of numeric values.
```

Virginia is the highest birth state for U.S. Presidents.

Graph 3: Total Vehicle Sales

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
df_tvs = pd.read_csv('TOTALNSA.csv')
```

```python
df_tvs[["DATE"]] = df_tvs[["DATE"]].apply(pd.to_datetime) #Changes 'Date' column type from an object to datetime.
```

```python
fig,ax = plt.subplots()
ax.plot(df_tvs['DATE'], df_tvs['TOTALNSA'])
fig.suptitle('Total Vehicle Sales')   # Graph title
ax.xaxis.set_label_text('Date')  # x-axis caption
ax.yaxis.set_label_text('Vehicle Sales per Month')  #y-axis caption 
fig.set_size_inches(15,5)
plt.show()

#I chose a regular time series plot because this is for time series data. 
```

U.S. vehicle sales have varied seasonally and cyclically.
